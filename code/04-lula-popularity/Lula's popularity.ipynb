```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.pylab as pylab
from matplotlib.ticker import FuncFormatter


%matplotlib inline
```

```python
df = pd.read_csv('popularidade-lula.csv')
df
```

```python
df['data_formatada'] = pd.to_datetime(df['data'], format="%m/%Y") 
df.head()
```

```python
df.drop(df.head(4).index, inplace=True)
```

```python
df.head()
```

```python
fig, ax = plt.subplots()

ax = df.desaprova.plot(x='data_formatada', y='desaprova', color='#dd3e3e', legend=True, figsize=(15,12))
df.aprova.plot(x='data_formatada', y='aprova', ax=ax, color='#17139c', legend=True)
df.ns.plot(x='data_formatada', y='ns', ax=ax, color='#8787a3', legend=True)

ax.set_xticklabels(['', 'May 2016', 'Oct 2016', 'Mar 2017', 'Aug 2017', 'Jan 2018', 'Jun 2018'])
ax.set_yticks([0, 25, 50, 75])

ax.set_title('Lula\'s popularity over the last two years', pad=50)

ax.title.set_fontsize(25)
ax.title.set_fontname('Franklin Gothic Heavy')

ax.spines['top'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['right'].set_visible(False)

ax.tick_params(which='both', left=False, bottom=True, grid_alpha=0.9)

ax.grid(True, axis='y', color='#febfb3', linestyle='--', linewidth=0.8)
ax.set_axisbelow(True)

plt.legend(bbox_to_anchor=(0.9, 1.1), loc=2, borderaxespad=0.)

plt.rcParams.update({
    'pdf.fonttype': 42,
    'ps.fonttype': 42,
    'figure.facecolor': '#f2f2f2', # colors the outside
    'axes.facecolor': '#f2f2f2', # colors the inside 
    'font.size': 12,
    'text.color': '#000000',
    'axes.labelcolor': '#ffdbc5',
    'legend.fancybox': True
})

plt.savefig("lula-popularity.pdf")
```
